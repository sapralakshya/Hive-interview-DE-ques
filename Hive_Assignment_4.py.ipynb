# 1.  Write a Python program that uses the HiveQL language to create a table named "Employees" with columns for "id," "name," and "salary."

```python
from pyhive import hive

def create_employees_table():
    # Establish a connection to Hive
    connection = hive.Connection(host="localhost", port=10000, auth="NOSASL")
    
    # Create a cursor object
    cursor = connection.cursor()
    
    # Execute the HiveQL query to create the table
    query = """
    CREATE TABLE Employees (
        id INT,
        name STRING,
        salary DOUBLE
    )
    """
    cursor.execute(query)
    
    # Close the cursor and connection
    cursor.close()
    connection.close()
    
    print("Employees table created successfully!")

# Call the function to create the Employees table
create_employees_table()
```

# 2.  Create a Python program that retrieves records from a Hive table named "Customers" where the age is greater than 30.

```python
from pyhive import hive

def retrieve_customers():
    # Establish a connection to Hive
    connection = hive.Connection(host="localhost", port=10000, auth="NOSASL")
    
    # Create a cursor object
    cursor = connection.cursor()
    
    # Execute the HiveQL query to retrieve records
    query = """
    SELECT * FROM Customers WHERE age > 30
    """
    cursor.execute(query)
    
    # Fetch all the records
    records = cursor.fetchall()
    
    # Print the retrieved records
    print("Customers with age greater than 30:")
    for record in records:
        print(record)
    
    # Close the cursor and connection
    cursor.close()
    connection.close()

# Call the function to retrieve the customers
retrieve_customers()
```

# 3.  Write a Python script that sorts records in descending order based on the "timestamp" column in a Hive table named "Logs."

```python
from pyhive import hive

def sort_logs_by_timestamp():
    # Establish a connection to Hive
    connection = hive.Connection(host="localhost", port=10000, auth="NOSASL")
    
    # Create a cursor object
    cursor = connection.cursor()
    
    # Execute the HiveQL query to sort records
    query = """
    SELECT * FROM Logs ORDER BY timestamp DESC
    """
    cursor.execute(query)
    
    # Fetch all the sorted records
    sorted_records = cursor.fetchall()
    
    # Print the sorted records
    print("Logs sorted by timestamp (descending order):")
    for record in sorted_records:
        print(record)
    
    # Close the cursor and connection
    cursor.close()
    connection.close()

# Call the function to sort the logs
sort_logs_by_timestamp()
```

# 5.  Write a Python script that calculates the average salary of employees from a Hive table named "Employees".

```python
from pyhive import hive

def calculate_average_salary():
    # Establish a connection to Hive
    connection = hive.Connection(host="localhost", port=10000, auth="NOSASL")
    
    # Create a cursor object
    cursor = connection.cursor()
    
    # Execute the HiveQL query to calculate average salary
    query = """
    SELECT AVG(salary) FROM Employees
    """
    cursor.execute(query)
    
    # Fetch the result of the query
    result = cursor.fetchone()
    
    # Extract the average salary from the result
    average_salary = result[0]
    
    # Print the average salary
    print("Average salary of employees:", average_salary)
    
    # Close the cursor and connection
    cursor.close()
    connection.close()

# Call the function to calculate the average salary
calculate_average_salary()
```

# 6.  Implement a Python program that uses Hive partitioning to create a partitioned table named "Sales_Data" based on the "year" and "month" columns.

```python
from pyhive import hive

def create_partitioned_table():
    # Establish a connection to Hive
    connection = hive.Connection(host="localhost", port=10000, auth="NOSASL")
    
    # Create a cursor object
    cursor = connection.cursor()
    
    # Execute the HiveQL query to create the partitioned table
    query = """
    CREATE TABLE Sales_Data (
        id INT,
        product_name STRING,
        sales DOUBLE
    )
    PARTITIONED BY (year INT, month INT)
    """
    cursor.execute(query)
    
    # Close the cursor and connection
    cursor.close()
    connection.close()
    
    print("Partitioned table Sales_Data created successfully!")

# Call the function to create the partitioned table
create_partitioned_table()
```

# 8.  Create a Python program that performs an inner join between two Hive tables, "Orders" and "Customers," based on a common column.

```python
from pyhive import hive

def perform_inner_join():
    # Establish a connection to Hive
    connection = hive.Connection(host="localhost", port=10000, auth="NOSASL")
    
    # Create a cursor object
    cursor = connection.cursor()
    
    # Execute the HiveQL query to perform inner join
    query = """
    SELECT *
    FROM Orders
    INNER JOIN Customers
    ON Orders.customer_id = Customers.customer_id
    """
    cursor.execute(query)
    
    # Fetch all the joined records
    joined_records = cursor.fetchall()
    
    # Print the joined records
    print("Joined records from Orders and Customers tables:")
    for record in joined_records:
        print(record)
    
    # Close the cursor and connection
    cursor.close()
    connection.close()

# Call the function to perform the inner join
perform_inner_join()
```

# 9.  Implement a Python program that uses the Hive SerDe library to process JSON data stored in a Hive table named "User_Activity_Logs."

```python
from pyhive import hive
import json

def process_user_activity_logs():
    # Establish a connection to Hive
    connection = hive.Connection(host="localhost", port=10000, auth="NOSASL")
    
    # Create a cursor object
    cursor = connection.cursor()
    
    # Execute the HiveQL query to select JSON data from the table
    query = """
    SELECT *
    FROM User_Activity_Logs
    """
    cursor.execute(query)
    
    # Fetch all the records
    records = cursor.fetchall()
    
    # Process the JSON data
    print("Processed User Activity Logs:")
    for record in records:
        json_data = record[0]  # Assuming the JSON data is stored in the first column
        parsed_json = json.loads(json_data)
        print(parsed_json)
    
    # Close the cursor and connection
    cursor.close()
    connection.close()

# Call the function to process the user activity logs
process_user_activity_logs()
```
